```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
random_generator = np.random.default_rng(11112020)
```

```python
random_series = random_generator.normal(size=50)
```

Plot

```python
plt.plot(random_series)
#plt.savefig('noise_series.png', dpi=300)
```

Auto Correlation

```python
sm.graphics.tsa.plot_acf(random_series)
plt.show()
```

```python
acorr_ljungbox(random_series).lb_pvalue.round(2).tolist()
```

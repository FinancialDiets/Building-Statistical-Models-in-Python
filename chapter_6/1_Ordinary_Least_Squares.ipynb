```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot - The relationship between x and y

error = np.random.normal(0,1,50)
x = np.linspace(0,10,50)
y = 2*x+3 +error

# plot the results
plt.figure(figsize = (10,8))
plt.scatter(x,y,color='blue',label='Actual Values')
plt.plot(x, 2*x + 3, color='red',label='Line of Best Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('The relationship between x and y')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Function to find beta_0_hat and beta_1_hat

def least_squares_method(x,y): 
    x_mean=x.mean() 
    y_mean=y.mean() 
    beta1 = ((x-x_mean)*(y-y_mean)).sum(axis=0)/ ((x-x.mean())**2).sum(axis=0) 
    beta0 = y_mean-(beta1*x_mean) 
    return beta0, beta1 
```

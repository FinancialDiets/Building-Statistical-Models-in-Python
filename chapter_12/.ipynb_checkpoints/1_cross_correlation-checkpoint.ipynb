```python
import requests
import zipfile

import numpy as np
import pandas as pd
import statsmodels.tsa.stattools as smt
import matplotlib.pyplot as plt
```

```python
# data from UCI Machine Learning Repository:
#   Dua, D. and Graff, C. (2019). 
#   UCI Machine Learning Repository [http://archive.ics.uci.edu/ml]. 
#   Irvine, CA: University of California, School of Information and Computer Science.

# code to download the data
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00501/PRSA2017_Data_20130301-20170228.zip'
file_name = url.split('/')[-1]
r = requests.get(url)
with open(file_name, 'wb') as fout:
    fout.write(r.content)

with zipfile.ZipFile(file_name, 'r') as zip_file:
    zip_file.extractall('./')
```

```python
def plot_ccf(series_1, series_2, n_lags):
    lag_limit = n_lags + 1
    ccfs = smt.ccf(series_2, series_1, adjusted=False)[::-1][:lag_limit]
    lags = -np.arange(ccfs.size)
    plt.stem(lags, ccfs)
    plt.axhline(1.96/np.sqrt(len(series_1)), color='black', ls='--')
    plt.axhline(-1.96/np.sqrt(len(series_1)), color='black', ls='--')
```

```python
size = 7*24
plot_ccf(df.PRES.iloc[:size], df.SO2.iloc[:size], 50)
```

```python
plot_ccf(df.PRES.iloc[:size], df.WSPM.iloc[:size], 50)
```

```python
plot_ccf(df.PRES.iloc[:size], df.NO2.iloc[:size], 50)
```
